```python
import os, configparser
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from bayes_opt import BayesianOptimization
from bayes_opt import UtilityFunction
```

```python
def mysql_knob_dataframe(knobs_path):
    config_len = len(os.listdir(knobs_path))
    cnf_parser = configparser.ConfigParser()
    pd_mysql = pd.DataFrame()
    for idx in range(config_len):
        cnf_parser.read(os.path.join(knobs_path, f'my_{idx}.cnf'))
        conf_dict = cnf_parser._sections['mysqld']
        tmp = pd.DataFrame(data=[conf_dict.values()], columns=conf_dict.keys())
        pd_mysql = pd.concat([pd_mysql, tmp])
        
    pd_mysql = pd_mysql.reset_index(drop=True)
    pd_mysql = pd_mysql.drop(columns=['log-error', 'bind-address', 'datadir', 'basedir'])
    return pd_mysql
```

```python
knobs = mysql_knob_dataframe("samples/ycsb_A/configs/").astype(float)
externals = pd.read_csv("samples/ycsb_A/results/external_metrics.csv", index_col=0)

plt.figure(figsize=(12, 5))
plt.plot(externals['tps'] / externals['latency'])
```

```python
def remove_inf_value(a:np.array) -> np.array:
    cnt = -1
    while(True):
        s_max_idx = np.argsort(a)[cnt]
        if a[s_max_idx] == np.inf:
            cnt -= 1
        else:
            break
    s_max_values = a[s_max_idx]
    
    return np.where(a==np.inf, s_max_values, a)
```

```python
x_array = np.array(knobs)
y_array = np.array(externals)
y_array[:,1] = remove_inf_value(y_array[:,1])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import sklearn.preprocessing as sklearn_scaler
```

```python
def fit_rfr(x_array, y_array, scaler=None):
    x_train, x_test, y_train, y_test = train_test_split(x_array,y_array, test_size=0.3, shuffle=True, random_state=11)

    if scaler is not None:
        print(f"Scaling with {scaler}")
        x_scaler = getattr(sklearn_scaler, scaler)()
        y_scaler = getattr(sklearn_scaler, scaler)()
        x_scaler.fit(x_train)
        y_scaler.fit(y_train)

        x_train = x_scaler.transform(x_train)
        x_test = x_scaler.transform(x_test)
        y_train = y_scaler.transform(y_train) 
        y_test = y_scaler.transform(y_test)
    else:
        print("Not scaling")
    
    regr = RandomForestRegressor(max_depth=4)
    regr.fit(x_train, y_train)

    print(regr.score(x_test, y_test))
    
    pred = regr.predict(x_test)
    if scaler is not None and scaler != 'Normalizer':
        pred = y_scaler.inverse_transform(pred)
        y_test = y_scaler.inverse_transform(y_test)
        
    plt.figure(figsize=(12, 5))
    plt.plot(y_test[:,0]/y_test[:,1], label='original')
    plt.plot(pred[:,0]/pred[:,1], label='estimates')
    plt.legend()
    plt.show()
```

```python
# scalers = [None, "StandardScaler", "Normalizer", "MinMaxScaler", "RobustScaler"]
scalers = [None, "StandardScaler", "MinMaxScaler", "RobustScaler"]
```

```python
fit_rfr(x_array, y_array, "Normalizer")
```

```python
for s in scalers:
    fit_rfr(x_array, y_array, s)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_array,y_array, test_size=0.3, shuffle=True, random_state=11)

scaler = "RobustScaler"
x_scaler = getattr(sklearn_scaler, scaler)()
y_scaler = getattr(sklearn_scaler, scaler)()
x_scaler.fit(x_train)
y_scaler.fit(y_train)

x_train = x_scaler.transform(x_train)
x_test = x_scaler.transform(x_test)
y_train = y_scaler.transform(y_train) 
y_test = y_scaler.transform(y_test)

regr = RandomForestRegressor(max_depth=4)
regr.fit(x_train, y_train)
```

```python
def obj(x):
    s_x = x_scaler.transform(x)
    pred = regr.predict(s_x)
    pred = y_scaler.inverse_transform(pred)
    res = pred[:,0]/pred[:,1]
    return res.squeeze()


def f(**kwargs):
    x = np.fromiter(kwargs.values(), dtype=float)
    
    s_x = x_scaler.transform([x])
    pred = regr.predict(s_x)
    pred = y_scaler.inverse_transform(pred)
    res = pred[:,0]/pred[:,1]
    return res.squeeze()

invs_y_test = y_scaler.inverse_transform(y_test)
plt.figure(figsize=(12, 5))
plt.plot(invs_y_test[:,0]/invs_y_test[:,1])
```

```python
pbounds = {}
for col in knobs:
    pbounds[col] = (knobs[col].min(), knobs[col].max())
```

```python
bo = BayesianOptimization(
    f=f,
    pbounds=pbounds,
    verbose=0,
    random_state=987234,
)

acquisition_function = UtilityFunction(kind="ucb", kappa=0.1)
bo.maximize(n_iter=10, acquisition_function=acquisition_function)
```
